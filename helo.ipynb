```python
while True:
    print('Calculator Menu')
    print('ADD\tSUB\tDIV\tMUL\tEXit')
    print('___'*20)
    print('>>')
    ch= input('Select Options')
    if ch.lower()=='add':
        x=int(input('Enter a:'))
        y=int(input('Enter b:'))
        print(f'{x} + {y} = {x+y}')
        print('___'*20)
    elif ch.lower()=='sub':
        x=int(input('Enter a:'))
        y=int(input('Enter b:'))
        print(f'{x} - {y} = {x-y}')
        print('___'*20)
    elif ch.lower()=='div':
        x=int(input('Enter a:'))
        y=int(input('Enter b:'))
        print(f'{x} / {y} = {x/y}')
        print('___'*20)
    elif ch.lower()=='mul':
        x=int(input('Enter a:'))
        y=int(input('Enter b:'))
        print(f'{x} * {y} = {x*y}')
        print('___'*20)
    elif ch.lower()=='exit':
        print('Calc closed')
        break

            
```
